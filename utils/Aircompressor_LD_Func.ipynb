```python
import numpy as np
import pandas as pd
import csv

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
def ReadTable(status = 'healthy', read_size = 225):
        for i in range(1,read_size + 1):
            if (i % 10 == 0):
                print("Read {} Files.".format(i))
            if i == 1:
                read_list = []
                read_file = open(f'../dataset/AirCompressor_Data/{status}/preprocess_Reading{i}.dat',mode='r',encoding='utf8', newline='\r\n')
                for line in read_file:
                    read_list.append(line.split(','))
                
                tmp_read_list = [float(idx) for idx in read_list[0]]
                first_file_series = pd.Series(tmp_read_list)
                
            elif i == 2 :
                tmp_read_list = []
                tmp_read_file = open(f'../dataset/AirCompressor_Data/{status}/preprocess_Reading{i}.dat',mode='r',encoding='utf8', newline='\r\n')
                
                for line in tmp_read_file:
                    tmp_read_list.append(line.split(','))
                    
                tmp_read_list = [float(idx) for idx in tmp_read_list[0]]
                second_file_series = pd.Series(tmp_read_list)
                
                df = pd.concat([first_file_series, second_file_series], axis = 1)
                
            else: # i >= 3
                tmp_read_list = []
                tmp_read_file = open(f'../dataset/AirCompressor_Data/{status}/preprocess_Reading{i}.dat',mode='r',encoding='utf8', newline='\r\n')
                for line in tmp_read_file:
                    tmp_read_list.append(line.split(','))
                
                tmp_list = [float(idx) for idx in tmp_read_list[0]]
                else_file_series = pd.Series(tmp_list)
                
                df = pd.concat([df, else_file_series], axis = 1)
                
                col_names = ["s{}".format(i) for i in range(1, df.shape[1] + 1)]
                df.columns = col_names
                
                if i == read_size:
                    print("Label: {} Dataset Load Completed".format(status))
                    
                    return df
```

```python
healthy = ReadTable(status = 'healthy')
liv = ReadTable(status = 'LIV')
```
